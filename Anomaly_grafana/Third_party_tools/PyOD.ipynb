```python
# Importing necessary libraries
from pyod.models.abod import ABOD
from pyod.models.knn import KNN
from pyod.models.iforest import IForest
from pyod.models.ocsvm import OCSVM
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the CSV file
file_path = '/content/calls_total_df.csv'
df = pd.read_csv(file_path)

# Extract the 'Value' column for outlier detection
X = df[['Value']].values

# Define the models to be used
models = {
    "ABOD": ABOD(),
    "KNN": KNN(),
    "IForest": IForest(),
    "OCSVM": OCSVM(),
}

# Fit the models and collect the results
results = {}

for name, model in models.items():
    model.fit(X)
    results[name] = model.decision_function(X)  # Higher scores indicate more abnormality

# Plotting the results
plt.figure(figsize=(14, 8))

for i, (name, decision_scores) in enumerate(results.items(), 1):
    plt.subplot(2, 2, i)
    plt.title(name)
    plt.scatter(np.arange(len(X)), X, c=decision_scores, cmap='RdYlBu', edgecolor='k', s=20)
    plt.colorbar(label='Anomaly Score')
    plt.xlabel('Instance')
    plt.ylabel('Value')

plt.tight_layout()
plt.show()
```
