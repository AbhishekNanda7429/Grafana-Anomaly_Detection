```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the CSV file
file_path = '/content/calls_total_df.csv'
df = pd.read_csv(file_path)

# Assuming the first column of CSV contains the datetime information
# Replace '0' with the actual column index if it's different
df['datetime'] = pd.to_datetime(df.iloc[:, 0]) # Changed line: Extracts the first column (index 0) of the DataFrame and converts it to datetime
df.set_index('datetime', inplace=True)

#-----------------------------------
# window_size = 10  # Adjust the window size depending on the frequency of your data
# df['rolling_mean'] = df['Value'].rolling(window=window_size).mean()
# df['rolling_std'] = df['Value'].rolling(window=window_size).std()

# df['z_score'] = (df['Value'] - df['rolling_mean']) / df['rolling_std']
# df['anomaly'] = np.where(np.abs(df['z_score']) > 3, 1, 0)
#-----------------------------------

decomposition = seasonal_decompose(df['Value'], model='additive', period=1)  # Adjust period based on your data
df['detrended'] = df['Value'] - decomposition.trend
df['deseasonalized'] = df['detrended'] - decomposition.seasonal

df['z_score'] = zscore(df['deseasonalized'].dropna())
df['anomaly'] = np.where(np.abs(df['z_score']) > 2.5, 1, 0)

# Plotting the data
plt.figure(figsize=(15, 8))
plt.plot(df.index, df['Value'], label='Value') # Changed line: Replace 'Value' with the actual column name
plt.scatter(df.index[df['anomaly'] == 1], df['Value'][df['anomaly'] == 1], color='red', label='Anomaly') # Changed line: Replace 'Value' with the actual column name
plt.title('Skyline Anomaly Detection')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()

# Optionally save the result to a new CSV
# output_path = '/mnt/data/skyline_anomalies.csv'
# df.to_csv(output_path)

# print(f"Anomaly detection results saved to {output_path}")
```
