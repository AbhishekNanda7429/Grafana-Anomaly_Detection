```python
import pandas as pd
import matplotlib.pyplot as plt
from adtk.data import validate_series
from adtk.visualization import plot
from adtk.detector import ThresholdAD

# Load the data
file_path = '/content/calls_total_df.csv'
df = pd.read_csv(file_path)

# Convert Time column to datetime and set as index
df['Time'] = pd.to_datetime(df['Time'])
df.set_index('Time', inplace=True)

# Validate series for anomaly detection
df_series = validate_series(df['Value'])

# Set up the threshold-based anomaly detector
threshold_ad = ThresholdAD(high=df_series.mean() + 3 * df_series.std(), low=df_series.mean() - 3 * df_series.std())

# Detect anomalies
anomalies = threshold_ad.detect(df_series)

# Plot the results
plt.figure(figsize=(12, 6))
plot(df_series, anomaly=anomalies, anomaly_color='red', anomaly_tag='marker')
plt.title('Anomaly Detection using ADTK')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```
